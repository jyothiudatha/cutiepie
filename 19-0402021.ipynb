```python
from random import randint
def marksgen(n,lb,ub):
    with open('datafiles/marks.txt','w') as f:
        for i in range(0,n):
            r=randint(lb,ub)
            f.write(str(r)+'\n')
    return    
n=int(input("enter student count"))
lb=int(input("enter lower range"))
ub=int(input("enter upper range"))
marksgen(n,lb,ub)
```

```python
#class average
with open('datafiles/marks.txt') as f:
    data=f.read()
    marks=data.split()
    sum=0
    for mark in marks:
        sum+=int(mark)
        print(sum)
    print(sum/len(marks))    
```

```python
#pass percentage
with open('datafiles/marks.txt') as f:
    data=f.read()
    marks=data.split()
    pc=0
    for marks in marks:
        if int(marks)>=35:
            pc+=1
    print(pc/len(marks)*100)        
```

```python
#least marks
print(marks.count(max(marks)))
print(marks.count(min(marks)))
```

```python
class A:
    def factorial(self,a):
        fact=1
        for i in range(1,a+1):
            fact=fact*i
        print("factorial:",fact)
a=A()
a.factorial(5)
```

```python
class A:
    def perfect(self,a):
        sum=0
        for i in range(1,a):
            if a%i==0:
                sum+=i
        if sum==a:
            return "it is a perfect number"
        else:
            return "not a perfect number"
a=A()
n=int(input("enter a number"))
a.perfect(n)
```

```python
class A:
    def prime(self,a):
        count=0
        for i in range(1,a+1):
            if a%i==0:
                count+=1
        if count==2:
            return "it is a prime number"
        else:
            return "not a prime number"
a=A()
a.prime(25)
        
        
```

```python
#single inheritance
class A():
    a=1
    def ca():
        print('from parent class')
class B(A):
    b=2
    def cb():
        print('from child class')
obj=B
print(obj.cb())
print(obj.ca())
        
        
```

```python
class A():
    a=1
    def ca():
        print('from parent class')
class B():
    b=2
    def cb():
        print('from parent class')
class C(A,B):
    c=3
    def cc():
        print('from child class')
obj=C
print(obj.ca())
print(obj.cb())
print(obj.cc())
        
```
